```python
import random

grid = [['-', '-', '-'],
        ['-', '-', '-'],
        ['-', '-', '-']]

def is_filled(grid):
  count = 0
  for i in range (3):
    for j in range(3):
      if grid[i][j] == '-':
        count += 1
  if (count == 0):
    return True
  else:
    return False

def has_won(grid, xo):
  if(grid[0][0] == grid[0][1] == grid[0][2] == xo or grid[1][0] == grid[1][1] == grid[1][2] == xo or grid[2][0] == grid[2][1] == grid[2][2] == xo or grid[0][0] == grid[1][0] == grid[2][0] == xo or grid[0][1] == grid[1][1] == grid[2][1] == xo or grid[0][2] == grid[1][2] == grid[2][2] == xo or grid[0][0] == grid[1][1] == grid[2][2] == xo or grid[2][0] == grid[1][1] == grid[0][2] == xo):
    return True
  else:
    return False

def show_board(grid):
  for i in range(3):
    for j in range(3):
      print(grid[i][j], end = " ")
    print()
  
#main body of program
#pick the Player who starts the game using random module
random.seed()
x = random.randint(3, 15)
start = x%2
xo = '-'
if(start == 1):
  print("Player 1 starts the game!")
  xo = 'X'
else:
  print("Player 2 starts the game!")
  xo = 'O'
once = False

while(True):
  show_board(grid)
  if(once == True):
    if(xo == 'X'):
      print("Player 2 takes turn...")
      xo = 'O'
    elif(xo == "O"):
      print("Player 1 takes turn...")
      xo = 'X'

  x = int(input("Enter row number(0-2): "))
  y = int(input("Enter column number(0-2): "))
  grid[x][y] = xo
  once = True
  if(has_won(grid, xo) == True):
    if(xo == 'X'):
      print("Player 1 wins!")
    elif(xo == 'O'):
      print("Player 2 wins!")
    break
  if(is_filled(grid) == True):
    print("It's a draw!")
    break

show_board(grid)
```
